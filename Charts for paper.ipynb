```python
# Set the random seed for reproducibility
np.random.seed(42)

# Generate a time series
time = np.arange(0, 100, 0.1)
seasonal_pattern = np.sin(time * 2 * np.pi / 12)  # seasonal pattern with a period of 12
trend = 0.05 * time  # linear trend
noise = np.random.normal(0, 0.5, len(time))  # random noise

# Combine the components to create a seasonal time series
seasonal_timeseries = 10 + trend + seasonal_pattern + noise

# Plot the seasonal time series
plt.figure(figsize=(10, 6))
plt.plot(time, seasonal_timeseries, color='green', linewidth=1)
plt.xlabel('Time', fontsize=15, fontname='Helvetica')
plt.ylabel('Value', fontsize=15, fontname='Helvetica')
plt.grid(True)
plt.savefig('seasonal-example.pdf')
plt.show()
```

```python
# Set the random seed for reproducibility
np.random.seed(42)

# Generate a time series
time = np.arange(15, 100, 0.1)
non_linear_trend = 0.05 * time**2  # non-linear trend (quadratic)
noise = np.random.normal(0, 15, len(time))  # random noise

# Combine the components to create a non-seasonal, non-linear time series
non_seasonal_timeseries = 10 + non_linear_trend + noise

# Plot the non-seasonal, non-linear time series
plt.figure(figsize=(10, 6))
plt.plot(time, non_seasonal_timeseries, color='green', linewidth=1)
plt.xlabel('Time', fontsize=15, fontname='Helvetica')
plt.ylabel('Value', fontsize=15, fontname='Helvetica')
plt.grid(True)
plt.savefig('non-seasonal-non-linear-example.pdf')
plt.show()
```

```python
# Generate a strictly stationary time series
time = np.arange(0, 100, 0.1)
noise = np.random.normal(0, 10, len(time))  # random noise

# The stationary time series is simply the noise itself
stationary_timeseries = noise + 50

# Plot the strictly stationary time series
plt.figure(figsize=(10, 6))
plt.plot(time, stationary_timeseries, color='green', linewidth=1)
plt.xlabel('Time', fontsize=15, fontname='Helvetica')
plt.ylabel('Value', fontsize=15, fontname='Helvetica')
plt.grid(True)
plt.savefig('strictly-stationary-example.pdf')
plt.show()
```
